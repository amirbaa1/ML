## Modeling

*   we use scikit-learn perceptron algorithm MLP Classifier in package neural_network.
*   Train and Test in dataframe train_data 

* hidden_layer_sizes about the number of neurons in the hidden layers of the neural network. we set the number of neuron layers to 200.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

x,y = train_data.loc[:,train_data.columns!='label'],train_data.loc[:,'label']

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.3,random_state=313)

model = MLPClassifier(hidden_layer_sizes=(200,), max_iter=100)
model.fit(x_train,y_train)
```

## Assessment

```python
from sklearn.metrics import classification_report, f1_score
y_pre = model.predict(x_test)
print(classification_report(y_test,y_pre))
print('f1 score :',f1_score(y_test,y_pre,average='weighted'))
```

* score accuracy 95%
