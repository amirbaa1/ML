```python
import fastf1
import pandas as pd
import numpy as np
import fastf1.plotting
from matplotlib import pyplot as plt
import matplotlib as mpl
from matplotlib.collections import LineCollection
```

Monaco race 2022 and 2023 P2

```python
race_2022 = fastf1.get_session(2022,'monaco','FP2')
race_2023 = fastf1.get_session(2023,'moncao','FP2')
```

```python
race_2022.load()
```

```python
race_2023.load()
```

```python
car_info_HAM_2022 = race_2022.laps.pick_driver('HAM').pick_fastest()
car_info_HAM_fast_2022 = car_info_HAM_2022.get_car_data()
car_info_HAM_fast_2022
```

```python
car_info_HAM_2023 = race_2023.laps.pick_driver('HAM').pick_fastest()
car_info_HAM_fast_2023 = car_info_HAM_2023.get_car_data()
car_info_HAM_fast_2023
```

```python
t = car_info_HAM_fast_2023['Time']
vCar = car_info_HAM_fast_2023['Speed']

t2 = car_info_HAM_fast_2022['Time']
vCar2 = car_info_HAM_fast_2022['Speed']
fastf1.plotting.setup_mpl(misc_mpl_mods=False)


fig, ax = plt.subplots(figsize=(20,10))
ax.plot(t, vCar, label='HAM 2023')
ax.plot(t2,vCar2,label = 'HAM 2022')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.set_title('HAM')
ax.legend()
plt.grid(True)
# invert the y-axis so drivers that finish higher are closer to the top
plt.show()
```

```python
maxspeed_2022= car_info_HAM_fast_2022['Speed'].max()
maxspeed_2023= car_info_HAM_fast_2023['Speed'].max()
```

```python
fig, ax = plt.subplots()

years = ['2022', '2023']
colors = ['#687BF3','#DDD000']
maxspeeds = [maxspeed_2022, maxspeed_2023]

bars = ax.bar(years, maxspeeds,color = colors)

for i, bar in enumerate(bars):
    ax.annotate(str(maxspeeds[i]), xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext=(0, 3), textcoords="offset points",
                ha='center', va='bottom')


ax.set_xlabel('Years')

ax.set_ylabel('Max Speed')

plt.show()
```

```python
b_2023 = car_info_HAM_fast_2023['Brake']
b_2022 = car_info_HAM_fast_2022['Brake']


fig,ax=plt.subplots(figsize=(20,3))

ax.plot(b_2023,label='HAM 2023')
ax.plot(b_2022,label='HAM 2022')
ax.set_xlabel('Brake')

ax.legend()
plt.show()
```

map

```python
weekend = race_2022.event
year = 2022
ses = 'FP2'
driver = 'HAM'
colormap = mpl.cm.plasma



x_2022 = car_info_HAM_2022.telemetry['X']
y_2022 = car_info_HAM_2022.telemetry['Y']
color_2022 = car_info_HAM_2022.telemetry['Speed']


points = np.array([x_2022, y_2022]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

```

```python
fig, ax = plt.subplots(sharex=True, sharey=True, figsize=(12, 6.75))
fig.suptitle(f'{weekend.name} {year} - {driver} - Speed', size=24, y=0.97)

# Adjust margins and turn of axis
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.12)
ax.axis('off')


# After this, we plot the data itself.
# Create background track line
ax.plot(x_2022, y_2022, color='black', linestyle='-', linewidth=16, zorder=0)

# Create a continuous norm to map from data points to colors
norm = plt.Normalize(color_2022.min(), color_2022.max())
lc = LineCollection(segments, cmap=colormap, norm=norm, linestyle='-', linewidth=5)

# Set the values used for colormapping
lc.set_array(color_2022)

# Merge all line segments together
line = ax.add_collection(lc)


# Finally, we create a color bar as a legend.
cbaxes = fig.add_axes([0.25, 0.05, 0.5, 0.05])
normlegend = mpl.colors.Normalize(vmin=color_2022.min(), vmax=color_2022.max())
legend = mpl.colorbar.ColorbarBase(cbaxes, norm=normlegend, cmap=colormap, orientation="horizontal")


# Show the plot
plt.show()
```
